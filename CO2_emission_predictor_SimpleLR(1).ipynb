# CO2 EMISSION PREDICTOR USING SIMPLE LINEAR REGRESSION

## Reading the Dataset using Pandas

```python
import pandas as pd

df=pd.read_csv("my2023-fuel-consumption-ratings.csv",encoding="latin")
data=df[["Engine size (L)","CO2 emissions (g/km)"]]
data
```

## Cleaning and Preprocessing

```python
data.info()  # no null cells in data no need to preprocess
```

```python
data.isnull().count()
```

## Finding relation between X and y

Using matplotlib create a graph to identify relation btw X and y and check if it is linear 


```python
import matplotlib.pyplot as plt

X=data['Engine size (L)']
y=data['CO2 emissions (g/km)']


plt.scatter(X,y,color='green')
plt.title("EngineSize v/s CO2 Emission")
plt.xlabel("engineSize")
plt.ylabel("co2Emission")
plt.show()
```

Here Relation is Linear as we can see , now we want to identify best fitted Line 
#### y = Q0 + Q1*X


## Training and Testing Data split

```python
from sklearn.model_selection import train_test_split

X=data[['Engine size (L)']]
y=data[['CO2 emissions (g/km)']]

train_X,test_X,train_y,test_y= train_test_split(X,y,test_size=0.22)
```

## Model Creation and Fitted Line detection


```python
from sklearn.linear_model import LinearRegression
import numpy as np

model = LinearRegression()

#fiting
model.fit(train_X,train_y)

print(f"Equation is y = {np.round(model.intercept_[0])} + {np.round(model.coef_[0][0])} * X")

predicted_y=model.predict(test_X)

```

```python

plt.scatter(train_X,train_y,color='green')
plt.plot(train_X,np.round(model.intercept_[0]) + np.round(model.coef_[0][0]) * train_X,color='red')
plt.title("EngineSize v/s CO2 Emission with Fitted Line")
plt.xlabel("engineSize")
plt.ylabel("co2Emission")
plt.show()
```

## Accuracy

#### if r2_score is
##### 0.7+     ==> very good fit
##### 0.5-0.7  ==> Acceptable and good fit
##### 0.5-     ==> Not a good fit , try another fitted line

```python
import numpy as np
from sklearn.metrics import r2_score 



print("Mean Absolute Error = ",np.mean(test_y-predicted_y))

print("Mean Square Error = ",np.mean((test_y-predicted_y)**2))

print("R2 Score = ",r2_score(test_y,predicted_y))


```

## RESULT

## Our Model's R2 Score is around 0.67 so it is a GOOD FITTED MODEL
