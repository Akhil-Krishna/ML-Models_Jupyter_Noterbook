# CO2 EMISSION OF CARS - MULTIPLE LINEAR REGRESSION

# Reading dataset using Pandas

```python
import pandas as pd

df= pd.read_csv('my2023-fuel-consumption-ratings.csv',encoding='latin1')

data=df[['Engine size (L)','Cylinders','CO2 emissions (g/km)']]
data
X=df[['Engine size (L)','Cylinders']]
y=df['CO2 emissions (g/km)']
```

# Plotting Scatter plots showing relation b/w Xs and y

```python
import matplotlib.pyplot as plt

plt.scatter(df['Engine size (L)'],df['CO2 emissions (g/km)'],color='blue')
plt.title('Engine Size vs CO2 emission')
plt.xlabel('Engine Size')
plt.ylabel('CO2 emission')
plt.show()

plt.scatter(df['Cylinders'],df['CO2 emissions (g/km)'],color='red')
plt.title('Cylinder vs CO2 emission')
plt.xlabel('Cylinder')
plt.ylabel('CO2 emission')
plt.show()
```

# Creating Model and its Equation

```python
import  numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score


train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.2)

model=LinearRegression()
model.fit(train_x,train_y)


print(f"y= Q0 + Q1x1 + Q2x2 ----->    y = {np.round(model.intercept_)} + {np.round(model.coef_[0])}X1 + {np.round(model.coef_[1])}X2")
```

# Plotting Edges of Hyper plain 

###### Not actual plot only ploting edges thats why lines donot matches with scatter plot 
###### Simply because it is a Multiple Linear Regression


```python
plt.scatter(df['Engine size (L)'],df['CO2 emissions (g/km)'],color='blue')
plt.plot(train_x['Engine size (L)'],np.round(model.intercept_) + np.round(model.coef_[0])*train_x['Engine size (L)'])
plt.title('Engine Size vs CO2 emission')
plt.xlabel('Engine Size')
plt.ylabel('CO2 emission')
plt.show()


plt.scatter(df['Cylinders'],df['CO2 emissions (g/km)'],color='red')
plt.plot(train_x['Cylinders'],np.round(model.intercept_) + np.round(model.coef_[0])*train_x['Cylinders'])
plt.title('Cylinder vs CO2 emission')
plt.xlabel('Cylinder')
plt.ylabel('CO2 emission')
plt.show()
```

# Predictiion and Accuracy

```python
predict_y=model.predict(test_x)


#MSE
print("Mean Squared Error = ",np.mean((test_y-predict_y)**2))

#MAE
print("Mean Absolute Error = ",np.mean(np.absolute(test_y-predict_y)))

#R squared Score

print("""\n\n --------R2_Score Chart--------
0.70-1.0  : Very good fit.
0.50-0.70 : Acceptable for many applications.
<0.50     : May indicate underfitting or missing variables.\n--------------------------\n""")
print("R2 Score = ",r2_score(test_y,predict_y))


```
